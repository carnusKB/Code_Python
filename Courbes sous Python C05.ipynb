```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import datetime
import matplotlib.dates as mdates
from matplotlib.transforms import Transform
from matplotlib.ticker import AutoLocator, AutoMinorLocator

fig, ax = plt.subplots(constrained_layout=True, facecolor=(.85, .87, .96))
t = np.arange(0, 360, 1)
s = 1.5 + np.sin(2 * t * np.pi / 180)
ax.set_facecolor('#eafff5')
ax.plot(t, s, 'r')
ax.set_xlabel('Angle [degrés]')
ax.set_ylabel('Amplitude [V]')
ax.set_title('Signal',fontsize='x-large')

def deg2rad(t):
    return t * np.pi / 180

def rad2deg(t):
    return t * 180 / np.pi

secax = ax.secondary_xaxis('top', functions=(deg2rad, rad2deg))
secax.set_xlabel('Angle [rad]')


def lin2log(s):
    return np.log10(s)

def log2lin(s):
    return 10**s

secay = ax.secondary_yaxis('right', functions=(lin2log, log2lin))
secay.set_ylabel('Log')

plt.show()
```

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

t = np.arange(0, 5.0+.01, .01)
data1 = t**2
data2 = np.sin(2 * np.pi * t)

fig, ax1 = plt.subplots()

color = 'tab:red'
ax1.set_xlabel('time (s)')
ax1.set_ylabel('exp', color=color)
ax1.plot(t, data1, color=color)
ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx()

color = 'tab:blue'
ax2.set_ylabel('sin', color=color)
ax2.plot(t, data2, color=color)
ax2.tick_params(axis='y', labelcolor=color)

fig.tight_layout()
plt.show()
```

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
t = np.arange(0, 5+0.05, 0.05)
y = t**2
z = -t**3

fig, ax1 = plt.subplots()

ax2 = ax1.twinx()

ax1.plot(t, y, 'm-')
ax2.plot(t, z, 'b-')
ax1.set_xlabel('Temps')
ax1.set_ylabel('Fonction y', color='m')
ax2.set_ylabel('Fonction Z', color='b')

fig.tight_layout()
plt.show()
```
