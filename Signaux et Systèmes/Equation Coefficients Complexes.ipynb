## <cite><font color="blue"> Résolution d'équations de 2nd ordre à coefficients complexes : </font></cite>


<div class="alert alert-block alert-info"> 
    
On considère l'équation de 2nd ordre à coefficients complexes : $Ap^2+Bp+C=0$, avec $A, \ B, \ C \in \mathbb{C}$ et $A \ne 0$.

On calcule de discriminant $\Delta=B^2-4AC$.

$\delta=a+jb$ est la racine carrée de $\Delta$ si :

$\left\{ \begin{array}{ll} 
a^2+b^2=|\Delta| & (1)\\
a^2-b^2=\Re e(\Delta) &  (2)\\
2ab=\Im m(\Delta) &  (3)\\
\end{array}
\right.$
$\quad \Rightarrow \left\{ \begin{array}{ll} 
2a^2=|\Delta|+\Re e(\Delta) & \\
2b^2=|\Delta|-\Re e(\Delta) &  \\
2ab=\Im m(\Delta) &  \\
\end{array}
\right.$
$\quad \Rightarrow \left\{ \begin{array}{ll} 
a=\pm\sqrt{\dfrac{|\Delta|+\Re e(\Delta)}{2}} & \\
b=\pm\sqrt{\dfrac{|\Delta|-\Re e(\Delta)}{2}} &  \\
2ab=\Im m(\Delta) &  \\
\end{array}
\right.$

$\Rightarrow$ on obtient 2 racines de $\Delta$.

$p_{1,2}=\dfrac{-B\pm \delta_i}{2A}$

- <cite><font color="red"> Exemple : </font></cite> $p_n^2+(0.1646+ j0.9336)p_n+1=0$

$\Rightarrow \Delta=(0.1646+ j0.9336)^2-4=-4.8445+j0.3072$.
$\Rightarrow |\Delta|=\sqrt{(-4.8445)^2+(0.3072)^2}=4.8542$.

On doit résoudre ce système d'équations pour trouver $\sqrt{\Delta}$.

$\left\{ \begin{array}{ll} %{\Large{\ding{172}}}
a^2+b^2=4.8542 & \\
a^2-b^2=-4.8445 & \\
2ab=0.3072>0 & \\
\end{array}
\right.$
$\Rightarrow \left\{ \begin{array}{ll}
 2a^2=0.0097 & \\
 2b^2=9.6987 & \\
 ab>0 & \\
\end{array}
\right.$
$\Rightarrow \left\{ \begin{array}{l}
a=\pm 0.0696 \\
b=\pm 2.2021 \\
ab>0 \\
\end{array}
\right.$

$a$ et $b$ doivent avoir le même signe.
$\Rightarrow \left\{ \begin{array}{l}
\delta_1=+0.0696+j2.2021 \\
\delta_2=-0.0696-j2.2021 \\
\end{array}
\right.$

$\Rightarrow \left\{ \begin{array}{l}
p_{1}=\dfrac{-(0.1646+ j0.9336)-(0.0696+j2.2021)}{2} \\
p'_{1}=\dfrac{-(0.1646+ j0.9336)+(0.0696+j2.2021)}{2} \\
p_{1}=\dfrac{-(0.1646+ j0.9336)-(-0.0696-j2.2021)}{2} \\
p'_{1}=\dfrac{-(0.1646+ j0.9336)+(-0.0696-j2.2021)}{2} \\
\end{array}
\right.$
$\Rightarrow \left\{ \begin{array}{l}
p_{1}=-0.1171-j1.56785 \\
p'_{1}=-0.0475+j0.63425 \\
\end{array}
\right.$

## Remarques :

<div class="alert alert-block alert-danger">
    
Les racines d'une équation de 2nd ordre <cite><font color="green"> à coefficients complexes </font></cite> ($Ap^2+Bp+C=0$) ne sont pas forcément complexes conjuguées (voir l'exemple précédent).

Dans le cas des filtres, $B=-\Delta \omega_n p_{_{LP_i}}$. Les $p_{_{LP_i}}$ utilisés pour calculer $B$ sont complexes conjugués. 

* A cet effet, quand on utilise $p_{_{LP_i}}$ pour calculer $B$, l'équation de 2nd ordre à coefficients complexes admet deux solutions complexes.

* Et quand on utilise le conjugué de $p_{_{LP_i}}$ ($p^*_{_{LP_i}}$) pour calculer $B$, l'équation de 2nd ordre à coefficients complexes admet deux solutions complexes qui sont les conjugués des deux solutions obtenues dans le premier cas.

</div>

## Démonstartion (cas de notre équation $p_n^2-\Delta \omega_n p_{_{LP_i}} p_n+1=0$ ) :

<div class="alert alert-block alert-warning">
    
On considère l'équation de 2nd ordre à coefficients complexes : $p_n^2-\Delta \omega_n p_{_{LP_i}} p_n+1=0$, avec $p_{_{LP_i}}=x+jy$ un pôle de notre filtre passe-bas normalisé.

On calcule le discriminant $\Delta_1=(-\Delta \omega_n p_{_{LP_i}})^2-4=(-\Delta \omega_n (x+jy))^2-4$.

$\Rightarrow \Delta_1=\Delta \omega_n^2(x^2-y^2+j2xy)-4=(\Delta\omega_n^2x^2-\Delta\omega_n^2y^2-4)+j(2xy\Delta\omega_n^2)$.

$\Rightarrow \left\{ \begin{array}{ll} 
 \Re e(\Delta_1)=\Delta\omega_n^2x^2-\Delta\omega_n^2y^2-4 &  \\
 \Im m(\Delta_1)=2xy\Delta\omega_n^2 &  \\
 |\Delta_1|=\sqrt{(\Delta\omega_n^2x^2-\Delta\omega_n^2y^2-4)^2+(2xy\Delta\omega_n^2)^2} &  \\
\end{array}
\right.$

$\delta=a_1+jb_1$ est la racine carrée de $\Delta_1$ si :

$\left\{ \begin{array}{ll} 
a_1^2+b_1^2=|\Delta_1| & (1)\\
a_1^2-b_1^2=\Re e(\Delta_1) &  (2)\\
2a_1b_1=\Im m(\Delta_1) &  (3)\\
\end{array}
\right.$
$\quad \Rightarrow \left\{ \begin{array}{ll} 
2a_1^2=|\Delta_1|+\Re e(\Delta_1) & \\
2b_1^2=|\Delta_1|-\Re e(\Delta_1) &  \\
2a_1b_1=\Im m(\Delta_1) &  \\
\end{array}
\right.$
$\quad \Rightarrow \left\{ \begin{array}{ll} 
a_1=\pm\sqrt{\dfrac{|\Delta_1|+\Re e(\Delta_1)}{2}} & \\
b_1=\pm\sqrt{\dfrac{|\Delta_1|-\Re e(\Delta_1)}{2}} &  \\
2a_1b_1=\Im m(\Delta_1) &  \\
\end{array}
\right.$

$\Rightarrow$ on obtient 2 racines de $\Delta_1$ : $\delta_1$ et $\delta_2$.

$p_{1,2}=\dfrac{\Delta \omega_n p_{_{LP_i}}\pm \delta_i}{2}=\dfrac{\Delta \omega_n (x+jy)\pm \delta_i}{2}$

---
Maintenant, on considère l'équation de 2nd ordre à coefficients complexes : $p_n^2-\Delta \omega_n p^*_{_{LP_i}} p_n+1=0$, avec $p^*_{_{LP_i}}$ est le conjugué de $p_{_{LP_i}}$. $\Rightarrow p^*_{_{LP_i}}=x-jy$.

On calcule le discriminant $\Delta_2=(-\Delta \omega_n p^*_{_{LP_i}})^2-4=(-\Delta \omega_n (x-jy))^2-4$.

$\Rightarrow \Delta_2=\Delta \omega_n^2(x^2-y^2-j2xy)-4=(\Delta\omega_n^2x^2-\Delta\omega_n^2y^2-4)-j(2xy\Delta\omega_n^2)$.

$\Rightarrow \left\{ \begin{array}{ll} 
 \Re e(\Delta_2)=\Delta\omega_n^2x^2-\Delta\omega_n^2y^2-4=\Re e (\Delta_1) &  \\
 \Im m(\Delta_2)=-2xy\Delta\omega_n^2=-\Im m(\Delta_1) &  \\
 |\Delta_2|=\sqrt{(\Delta\omega_n^2x^2-\Delta\omega_n^2y^2-4)^2+(-2xy\Delta\omega_n^2)^2}=|\Delta_1| &  \\
\end{array}
\right.$

$\delta'_2=a_2+jb_2$ est la racine carrée de $\Delta_2$ si :

$\left\{ \begin{array}{ll} 
a_2^2+b_2^2=|\Delta_2| & (4)\\
a_2^2-b_2^2=\Re e(\Delta_2) &  (5)\\
2a_2b_2=\Im m(\Delta_2) &  (6)\\
\end{array}
\right.$
$\quad \Rightarrow \left\{ \begin{array}{ll} 
2a_2^2=|\Delta_2|+\Re e(\Delta_2) & \\
2b_2^2=|\Delta_2|-\Re e(\Delta_2) &  \\
2a_2b_2=\Im m(\Delta_2) &  \\
\end{array}
\right.$
$\quad \Rightarrow \left\{ \begin{array}{ll} 
a_2=\pm\sqrt{\dfrac{|\Delta_2|+\Re e(\Delta_2)}{2}} & \\
b_2=\pm\sqrt{\dfrac{|\Delta_2|-\Re e(\Delta_2)}{2}} &  \\
2a_2b_2=\Im m(\Delta_2) &  \\
\end{array}
\right.$
$\quad \Rightarrow \left\{ \begin{array}{ll} 
a_2=\pm\sqrt{\dfrac{|\Delta_1|+\Re e(\Delta_1)}{2}} & \\
b_2=\pm\sqrt{\dfrac{|\Delta_1|-\Re e(\Delta_1)}{2}} &  \\
2a_2b_2=-\Im m(\Delta_1) &  \\
\end{array}
\right.$ 

Car $\quad \Rightarrow \left\{ \begin{array}{ll} 
\Re e(\Delta_2)=\Re e(\Delta_1) & \\
\Im m(\Delta_2)=-\Im m(\Delta_1) &  \\
|\Delta_2|=|\Delta_1| &  \\
\end{array}
\right.$

$\Rightarrow$ on obtient 2 racines de $\Delta_2$ : $\delta'_1$ et $\delta'_2$.

$p'_{1,2}=\dfrac{\Delta \omega_n p^*_{_{LP_i}}\pm \delta'_i}{2}=\dfrac{\Delta \omega_n (x-jy)\pm \delta'_i}{2}$

---
## <cite><font color="green"> Remarques : </font></cite>

On remarque que $|a_1|=|a_2|$ et $|b_1|=|b_2|$.

* Si $\Im m(\Delta_1)>0 \quad \Rightarrow \left\{ \begin{array}{ll} 
a_1=+\sqrt{\dfrac{|\Delta_1|+\Re e(\Delta_1)}{2}} & b_1=+\sqrt{\dfrac{|\Delta_1|+\Re e(\Delta_1)}{2}}\\
a_1=-\sqrt{\dfrac{|\Delta_1|-\Re e(\Delta_1)}{2}} & b_1=-\sqrt{\dfrac{|\Delta_1|-\Re e(\Delta_1)}{2}} \\
\end{array}
\right.$

$\Rightarrow \left\{ \begin{array}{ll} 
\delta_1=+\sqrt{\dfrac{|\Delta_1|+\Re e(\Delta_1)}{2}}+j\sqrt{\dfrac{|\Delta_1|+\Re e(\Delta_1)}{2}} & \\
\delta_2=-\sqrt{\dfrac{|\Delta_1|+\Re e(\Delta_1)}{2}}-j\sqrt{\dfrac{|\Delta_1|+\Re e(\Delta_1)}{2}} &  \\
\end{array}
\right.$

Donc $\left\{ \begin{array}{ll} 
\delta'_1=+\sqrt{\dfrac{|\Delta_1|+\Re e(\Delta_1)}{2}}-j\sqrt{\dfrac{|\Delta_1|+\Re e(\Delta_1)}{2}} & \\
\delta'_2=-\sqrt{\dfrac{|\Delta_1|+\Re e(\Delta_1)}{2}}+j\sqrt{\dfrac{|\Delta_1|+\Re e(\Delta_1)}{2}} &  \\
\end{array}
\right.$


* Si $\Im m(\Delta_1)<0 \quad \Rightarrow \left\{ \begin{array}{ll} 
a_1=+\sqrt{\dfrac{|\Delta_1|+\Re e(\Delta_1)}{2}} & b_1=-\sqrt{\dfrac{|\Delta_1|+\Re e(\Delta_1)}{2}}\\
a_1=-\sqrt{\dfrac{|\Delta_1|-\Re e(\Delta_1)}{2}} & b_1=+\sqrt{\dfrac{|\Delta_1|-\Re e(\Delta_1)}{2}} \\
\end{array}
\right.$

$\Rightarrow \left\{ \begin{array}{ll} 
\delta_1=+\sqrt{\dfrac{|\Delta_1|+\Re e(\Delta_1)}{2}}-j\sqrt{\dfrac{|\Delta_1|+\Re e(\Delta_1)}{2}} & \\
\delta_2=-\sqrt{\dfrac{|\Delta_1|+\Re e(\Delta_1)}{2}}+j\sqrt{\dfrac{|\Delta_1|+\Re e(\Delta_1)}{2}} &  \\
\end{array}
\right.$

Donc $\left\{ \begin{array}{ll} 
\delta'_1=+\sqrt{\dfrac{|\Delta_1|+\Re e(\Delta_1)}{2}}+j\sqrt{\dfrac{|\Delta_1|+\Re e(\Delta_1)}{2}} & \\
\delta'_2=-\sqrt{\dfrac{|\Delta_1|+\Re e(\Delta_1)}{2}}-j\sqrt{\dfrac{|\Delta_1|+\Re e(\Delta_1)}{2}} &  \\
\end{array}
\right.$

Dans les deux, on constate que $\delta_i$ et $\delta'_i$ sont des complexes conjugués ( $\delta'_1=\delta^*_1$ et $\delta'_2=\delta^*_2$)

Finalement $\left\{ \begin{array}{ll} 
p_{1,2}=\dfrac{\Delta \omega_n p_{_{LP_i}}\pm \delta_i}{2} & \\
p'_{1,2}=\dfrac{\Delta \omega_n p^*_{_{LP_i}}\pm \delta'_i}{2}=\dfrac{\Delta \omega_n p^*_{_{LP_i}}\pm \delta^*_i}{2}=p^*_{1,2} &  \\
\end{array}
\right.$

$\boxed{\text{Conclusion : On obtient deux paires de pôles complexes conjugués.}}$

### <cite><font color="red"> Note : </font></cite>

Dans le cas d'une cellule (étage) de premier ordre, le pôle est un réel négatif (pôle$=-FSF_i$) et l'équation devient : $p_n^2+FSF_i p_n+1=0$ avec $FSF_i\le1$ (voir les tables de Butterworth et Chebychev). Cette équation est une équation à coefficients réels.

Donc le discriminant $\Delta$ est forcément négatif. $\Rightarrow$ l'équation admet deux solutions <cite><font color="red"> complexes conjuguées </font></cite>.
