## Signaux analogiques

```python
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt

def rect(t,tau):    
    x = np.zeros(len(t))    
    for k,tk in enumerate(t):        
        if np.abs(tk) > tau/2.:            
            x[k] = 0        
        else:            
            x[k] = 1    
    return x
def tri(t,tau):    
    x = np.zeros(len(t))    
    for k,tk in enumerate(t):        
        if np.abs(tk) > tau/1.:            
            x[k] = 0        
        else:            
            x[k] = 1 - np.abs(tk)/tau    
    return x
def step(t,t0):    
    x = np.zeros(len(t))    
    for k,tt in enumerate(t):        
        if tt >= t0: # Saut à t0 (= 0 souvent)            
            x[k] = 1.0    
    return x
def rampe(t,tau):    
    x = np.zeros(len(t))    
    for k,tk in enumerate(t):        
        if tk < 0:            
            x[k] = 0        
        else:            
            x[k] = tk   
    return x
def rectangulaire(t):
    x = np.sign(np.sin(2*np.pi*t))
    return x

t = np.arange(-2,2+.001,.001)

plt.plot(t,rect(t,1))
plt.title('Signal rectangulaire')
plt.show()
plt.plot(t,tri(t,1))
plt.title('Signal triangulaire')
plt.show()
plt.plot(t,rect(t,.001))
plt.title('Impulsion de Dirac')
plt.show()
plt.plot(t,step(t,0))
plt.title('Echelon unité')
plt.show()
plt.plot(t,rampe(t,0))
plt.title('Rampe unité')
plt.show()

plt.plot(t,rectangulaire(t))
plt.title('Signal rectangulaire')
plt.show()
```

## Signaux numériques

```python
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt

def rect(t,tau):    
    x = np.zeros(len(t))    
    for k,tk in enumerate(t):        
        if np.abs(tk) > tau/2.:            
            x[k] = 0        
        else:            
            x[k] = 1    
    return x
def tri(t,tau):    
    x = np.zeros(len(t))    
    for k,tk in enumerate(t):        
        if np.abs(tk) > tau/1.:            
            x[k] = 0        
        else:            
            x[k] = 1 - np.abs(tk)/tau    
    return x
def step(t,t0):    
    x = np.zeros(len(t))    
    for k,tt in enumerate(t):        
        if tt >= t0: # Saut à t0 (= 0 souvent)            
            x[k] = 1.0    
    return x
def rampe(t,tau):    
    x = np.zeros(len(t))    
    for k,tk in enumerate(t):        
        if tk < 0:            
            x[k] = 0        
        else:            
            x[k] = tk   
    return x

t = np.arange(-10,11,1)

plt.stem(t,rect(t,6),use_line_collection=True)
plt.title('Signal rectangulaire')
plt.show()
plt.stem(t,tri(t,5),use_line_collection=True)
plt.title('Signal triangulaire')
plt.show()
plt.stem(t,rect(t,.001),use_line_collection=True)
plt.title('Impulsion de Dirac')
plt.show()
plt.stem(t,step(t,0),use_line_collection=True)
plt.title('Echelon unité')
plt.show()
plt.stem(t,rampe(t,0),use_line_collection=True)
plt.title('Rampe unité')
plt.show()
```

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def delta(n):
    if n == 0:
        return 1
    else:
        return 0

e_ = []
x_ = []
y_ = []
n = 10

for i in range(-n,n+1):
    e = delta(i)
    e_.append(e)

for i in range(-n,n+1):
    x = delta(i) + delta(i-1) + delta(i-2) + delta(i-3) + delta(i-4) + delta(i-5)
    x_.append(x)

for i in range(-n,n+1):
    y = - delta(i+5) + delta(i+2) + delta(i) + delta(i-1) - delta(i-4) + delta(i-6)
    y_.append(y)

plt.figure(1)
markerline, stemlines, baseline = plt.stem(range(-n,n+1),e_, '--',use_line_collection=True)
plt.setp(stemlines, 'color', 'b', 'linewidth', 2)
plt.setp(baseline, 'color', 'b', 'linewidth', 0.5)
plt.ylim([-2,2])
plt.xlim([-n,n])
plt.xlabel('$n$')
plt.ylabel('$e[n]$')
plt.title('Signal numérique 1')
plt.grid(True)

plt.figure(2)
markerline, stemlines, baseline = plt.stem(range(-n,n+1),x_, '--',use_line_collection=True)
plt.setp(stemlines, 'color', 'b', 'linewidth', 2)
plt.setp(baseline, 'color', 'b', 'linewidth', 0.5)
plt.ylim([-2,2])
plt.xlim([-n,n])
plt.xlabel('$n$')
plt.ylabel('$x[n]$')
plt.title('Signal numérique 2')
plt.grid(True)

plt.figure(3)
markerline, stemlines, baseline = plt.stem(range(-n,n+1),y_, '--',use_line_collection=True)
plt.setp(stemlines, 'color', 'b', 'linewidth', 2)
plt.setp(baseline, 'color', 'b', 'linewidth', 0.5)
plt.ylim([-2,2])
plt.xlim([-n,n])
plt.xlabel('$n$')
plt.ylabel('$x[n]$')
plt.title('Signal numérique 3')
plt.grid(True)

plt.show()
```
