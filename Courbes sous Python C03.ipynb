## Exemples de courbes

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

x_min = -0.1
x_max = 0.1

y_min = -1.5
y_max = 1.5

x = np.arange(-0.1, 0.1, 0.002)
y = 1.3*np.sin(20*np.pi*x+(np.pi/3))
z = 1.3*np.cos(20*np.pi*x+(np.pi/3))

plt.figure(figsize=(9,6))
plt.plot(x, y, 'k', label='sin')
plt.xlim(x_min, x_max)
plt.ylim(y_min, y_max)
# Ou plt.axis([x_min, x_max, y_min, y_max])
plt.plot(x, z, 'r--', linewidth=3.0, label='cos')
plt.xlabel('Abscisses')
plt.ylabel('Ordonnées\n(unité)')
plt.title('Signal' r'$\ s(t)$', fontsize=14)
plt.title('Titre à gauche', loc='left')
plt.title('Titre à droite', loc='right')
plt.text(.015, 1.2, r'$1.3\sin(\omega x+\pi/3)$')
plt.annotate('cosinus', xy=(-.03, 1.), xytext=(-.075, 1.),
             arrowprops=dict(facecolor='orange', shrink=0.05))
plt.annotate("", xy=(-np.pi/(3*20*np.pi), 0), xytext=(5*np.pi/(3*20*np.pi), 0),
             arrowprops=dict(arrowstyle="<->", connectionstyle="arc3"))
plt.text(2*np.pi/(3*20*np.pi), 0, 'Période : ' r'$\frac{2\pi}{\omega}$',
         {'color': 'black', 'fontsize': 20, 'ha': 'center', 'va': 'center',
          'bbox': dict(boxstyle="round", fc="white", ec="black", pad=0.2)})
plt.legend(title='Légendes :',loc='upper right',shadow=True, fontsize='x-large',
           facecolor='C1')
ax = plt.gca()
ax.set_xticklabels([]) # Pour supprimer les graduations des abscisses
th2 = plt.text(-.06, -.1, 'Rotation du texte', fontsize=14, rotation=45, 
               rotation_mode='anchor')
plt.grid()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

x = np.arange(0, 4*np.pi, 0.01)
y = np.sin(x)
z = np.sin(x+np.pi/4)
zm = np.ma.masked_where(abs(z) > 0.75, z)

lines = plt.plot(x, y, x, zm)
plt.setp(lines[0], linewidth=2)
plt.setp(lines[1], linewidth=3)

plt.legend(('Non masqué', 'Masqué si' r'$>\pm 0.75$'),
           loc='upper right')
plt.title('Courbe masquée sur un intervalle')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

ax = plt.subplot(111)
t = np.arange(.0, 2*np.pi+.01, .01)
for n in [1, 2, 3]:
    plt.plot(t, np.sin(n*t), label=r'$s_%d=sin(%dt)$'%(n,n))
ax.set(ylim=(-1.6,1.1))
leg = plt.legend(loc='best', ncol=2, mode="expand", shadow=True, fancybox=True)
leg.get_frame().set_alpha(0.5)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

t = np.arange(0, 3, .01)
s = np.sin(10*np.pi*t) * np.exp(-2*t)

plt.figure(figsize=(8,6))
plt.plot(t, s)

plt.axhline(y=-.8, linewidth=10, color='red', alpha=0.3)
plt.axhline(y=.1, xmin=.25, xmax=1, linestyle='--', color='#af1135', 
            linewidth='1')
plt.axhline(y=-.1, xmin=.25, xmax=1, linestyle='--', color='k', linewidth='1')
plt.axvline(x=1.07, linestyle='-.', color='C1')
plt.axvline(x=2.9, ymin=0, linewidth=10, color='#8d33ff')
plt.axhspan(-.1, .1, facecolor='0.5', alpha=0.5)
plt.axvspan(0, .75, facecolor='#11af76', alpha=0.4)
plt.axis([0, 3, -1, 1])

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

t = np.arange(0, 2.0+.001, .001)
s = np.sin(2*np.pi*t)

seuil_b = -.6
seuil_h = .6

pb = np.ma.masked_where(s > seuil_b, s)
pm = np.ma.masked_where((s < seuil_b) | (s > seuil_h), s)
ph = np.ma.masked_where(s < seuil_h, s)

fig, ax = plt.subplots()
ax.plot(t, pb, 'r', t, pm, t, ph, 'm')
plt.grid()
plt.show()
```
