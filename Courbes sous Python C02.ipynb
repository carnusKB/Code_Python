## Multiple Plot

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
t = np.linspace(0, 2*np.pi)
fig, ax = plt.subplots(2, 3, sharex='col', sharey='row')
ax[1,1].plot(t,np.cos(t))
ax[1,1].grid()
plt.show()
```

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def f(t):
    return np.sin(20*np.pi*t) * (1 + 0.75*np.cos(2*np.pi*t))

t = np.arange(0.0, np.pi, 0.01)

ax1 = plt.subplot(211)
ax1.margins(0.05)
ax1.plot(t, f(t))
ax1.set_title('Tracé normal')

ax2 = plt.subplot(223)
ax2.margins(.5, .5)
ax2.plot(t, f(t))
ax2.set_title('Tracé reculé')

ax3 = plt.subplot(224)
ax3.margins(x = -.1, y = -.3)
ax3.plot(t, f(t))
ax3.set_title('Tracé avancé')

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

t = np.arange(0.0, 2+0.005, 0.005)
p = np.sin(20*np.pi*t)
m = np.cos(2*np.pi*t)
s = p * (1 + 0.75*m)

plt.figure(figsize=(8,6))

plt.plot(t, s)
plt.axis([0, np.max(t), 1.2 * np.min(s), 2.6 * np.max(s)])
plt.xlabel('Temps')
plt.ylabel('Amplitude')
plt.title('Signal AM')

a = plt.axes([.65, .6, .2, .2], facecolor='k')
plt.plot(t, m)
plt.title('Message')
plt.xlim(0, np.max(t)/2)
plt.yticks([])

b = plt.axes([0.2, 0.6, .2, .2], facecolor='y')
plt.plot(t, p)
plt.title('Porteuse')
plt.xlim(0, np.max(t)/2)
plt.yticks([])

plt.show()
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

fig = plt.figure()
ax1 = fig.add_axes([0.1, 0.5, 0.8, 0.4],
                   xticklabels=[], ylim=(-2.2, 2.2))
ax2 = fig.add_axes([0.1, 0.1, 0.8, 0.4],
                   ylim=(-3.3, 3.3))

t = np.linspace(0, 2*np.pi)

ax1.plot(2*np.sin(t),'r')
ax2.plot(3*np.cos(2*t));
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(0, 2*np.pi)
y = np.sin(x)
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2)
ax1.plot(x, y)
ax2.plot(x, y**2, 'tab:orange')
ax3.plot(x, -y, 'tab:green')
ax4.plot(x, -y**2, 'tab:red')

for ax in fig.get_axes():
    ax.label_outer()

fig, (ax5, ax6) = plt.subplots(2)
ax5.plot(x, y)
ax6.plot(x + 1, -y)

fig, (ax7, ax8) = plt.subplots(2, sharex=True)
ax7.plot(x, y)
ax8.plot(x + 1, -y)

fig, axs = plt.subplots(3, sharex=True, sharey=True)#meme echelle
axs[0].plot(x, y ** 2)
axs[1].plot(x, 0.3 * y, 'r')
axs[2].plot(x, y, 'm')

plt.show()
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(0, 2*np.pi)
y = np.sin(x)
fig, axs = plt.subplots(2, 2, sharex='col', sharey='row',
                        gridspec_kw={'hspace': 0, 'wspace': 0})
(ax1, ax2), (ax3, ax4) = axs
ax1.plot(x, y)
ax2.plot(x, y**2, 'tab:orange')
ax3.plot(x + 1, -y, 'tab:green')
ax4.plot(x + 2, -y**2, 'tab:red')

plt.show()
```
